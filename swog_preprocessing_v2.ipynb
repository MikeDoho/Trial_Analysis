```python
# loading excel files
df_pub = pd.read_excel (r'SWOG_Database_Bhavani.xls','Publications')  # df_1
df_trial_abbrev = pd.read_excel (r'SWOG_Database_Bhavani.xls','Closed_Protocol_abbrev') # df_2 
df_trial_com = pd.read_excel (r'SWOG_Database_10_28_19.xlsx') #df_3

# dropping all the work bhavani did
df_trial_abbrev = df_trial_abbrev[df_trial_abbrev.columns.drop(list(df_trial_abbrev.filter(regex='Abstract')))]
df_trial_abbrev = df_trial_abbrev[df_trial_abbrev.columns.drop(list(df_trial_abbrev.filter(regex='Pub')))]

df_trial_abbrev = df_trial_abbrev[['Trial Number', 'Title', 'Activated', 'Closed', 'Accrual', 'Phase']]
df_trial_abbrev.head()
```

```python
# removing nonspecific STUDYID for publication list to be analyzed separately by hand
remove_STUDYID = ['MULTIPLE','OTHER','NONE']

for name in remove_STUDYID:
    df_pub = df_pub[df_pub['STUDYID']!=name]
df_pub.head()
```

```python
# filtering df_trial_abbrev; df_2

import re 
# x = "This is a sentence (lkajsdlfkj) (alkjsdfljasdf) [twice a day]"
# re.sub("[\(\[].*?[\)\]]", "", x)

#removing () and []
for i in range(len(df_trial_abbrev['Trial Number'])):
    try:
#         print(df_2['Trial Number'].iloc[i])
        df_trial_abbrev['Trial Number'].iloc[i] = re.sub("[\(\[].*?[\)\]]", "", df_trial_abbrev['Trial Number'].iloc[i]).replace(' ','')
        df_trial_abbrev['Trial Number'][i] = df_trial_abbrev['Trial Number'][i].replace('SWOG-','S')
    except (AttributeError,TypeError): pass
    
    # if starts with a letter and has a / then remove first half (xxx/yyy -> yyy)
#     print(i, df_2['Trial Number'].iloc[i])
    if isinstance(df_trial_abbrev['Trial Number'].iloc[i], int) == False:
        if len(df_trial_abbrev['Trial Number'].iloc[i].split('/',1)) > 1:
            df_trial_abbrev['Trial Number'].iloc[i] = df_trial_abbrev['Trial Number'].iloc[i].split('/',1)[1]
            
    else: pass
    
# remove space, ., and -
to_remove = ['.','-',' ']
for idd in range(len(df_trial_abbrev['Trial Number'])):
    for i in to_remove:
        if isinstance(df_trial_abbrev['Trial Number'].iloc[idd], int) == False:
            df_trial_abbrev['Trial Number'].iloc[idd] =  df_trial_abbrev['Trial Number'].iloc[idd].replace(i,'')
    

df_trial_abbrev.head(10)
  
```

```python
# filtering df_trial_com; 

import re 
# x = "This is a sentence (lkajsdlfkj) (alkjsdfljasdf) [twice a day]"
# re.sub("[\(\[].*?[\)\]]", "", x)

#removing () and []
for i in range(len(df_trial_com['Trial'])):
    try:

        df_trial_com['Trial'].iloc[i] = re.sub("[\(\[].*?[\)\]]", "", df_trial_com['Trial'].iloc[i]).replace(' ','')
        df_trial_com['Trial'][i] = df_trial_com['Trial'][i].replace('SWOG-','S')
    except (AttributeError,TypeError): pass
    
    # if starts with a letter and has a / then remove first half (xxx/yyy -> yyy)

    if isinstance(df_trial_com['Trial'].iloc[i], int) == False:
        if len(df_trial_com['Trial'].iloc[i].split('/',1)) > 1:
            df_trial_com['Trial'].iloc[i] = df_trial_com['Trial'].iloc[i].split('/',1)[1]
            
    else: pass
    
# remove space, ., and -
to_remove = ['.','-',' ']
for idd in range(len(df_trial_com['Trial'])):
    for i in to_remove:
        if isinstance(df_trial_com['Trial'].iloc[idd], int) == False:
            df_trial_com['Trial'].iloc[idd] =  df_trial_com['Trial'].iloc[idd].replace(i,'')
    
```

```python
# filtering df_pub

header_list = ['STUDYID','PUBTITLE','PUBNAME','PUBYEAR']
df_pub = df_pub.reindex(columns = header_list)



# remove space, ., and -
to_remove = ['.','-',' ']
for idd in range(len(df_pub.STUDYID)):
    for i in to_remove:
#         df_pub.loc[idd, 'STUDYID'] =  df_pub.loc[idd, 'STUDYID'].replace(i,'')
        df_pub.STUDYID.iloc[idd] =  df_pub.STUDYID.iloc[idd].replace(i,'')


#df_pub.loc[234, 'STUDYID']
    
```

```python
#header_list = ['Trial Number','Title']
header_list = ['STUDYID','Title','Activated','Closed','Phase']
df_trial_abbrev = df_trial_abbrev.rename(columns = {'Trial Number': 'STUDYID'})

header_list = ['STUDYID', 'Title', 'Activated', 'Closed', 'Accrual', 'Phase', 'Published', 'Disease']
df_trial_com = df_trial_com.rename(columns= {'Trial':'STUDYID', 'Disease Site/Research Committee':'Disease'})

df_trial_com = df_trial_com[['STUDYID', 'Title','Activated', 'Closed', 'Accrual', 'Phase', 'Published', 'Disease']]
```

```python
# copying dataframes to csv

def assess_dup(df):
    for d in df:
        print(f"total amount: {len(d)}")
        print(f"amount with duplicates: {len(d.drop_duplicates())}")
        print('')

assess_dup([df_pub, df_trial_abbrev, df_trial_com])

df_trial_abbrev = df_trial_abbrev.drop_duplicates()
df_pub = df_pub.drop_duplicates()
df_trial_com = df_trial_com.drop_duplicates()

df_trial_abbrev.to_csv('Trial_List.csv')
df_pub.to_csv('Pub_list.csv')
df_trial_com.to_csv('Trial_List_Complete.csv')
```
